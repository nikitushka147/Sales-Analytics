#Reporting.
Склади детальний звіт у файлі Google Colab із кодом, графіками та поясненнями. Зроби висновки з точки зору бізнес-логіки. Цей звіт можна буде використати у резюме як твій реальний кейс проєкту у Data Analytics.

```python
df = pd.merge(countries_df, events_df, on="country_code", how="right")
df = pd.merge(df, products_df, on ="product_id", how ="left")

df["shipping_days"] = (df["ship_date"] - df["order_date"]).dt.days
df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.to_period('M')
df['order_dayname'] = df['order_date'].dt.day_name()

df['revenue'] = df['units_sold'] * df['unit_price']
df['cost'] = df['units_sold'] * df['unit_cost']
df['profit'] = df['revenue'] - df['cost']
df['margin'] = df['profit'] / df['revenue']*100
df.head()
df.info()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

rev = df.groupby('item_type', observed=False)['revenue'].mean()
rev = rev.sort_values(ascending=False)
rev.plot(kind="bar", ax=ax[0, 0], color="skyblue")
df.groupby("item_type", observed=False)["revenue"].mean().plot(kind="bar", ax=ax[0, 0], color="skyblue")
ax[0, 0].set_title("Revenue by Category")  # Title of the subplot
ax[0, 0].set_xlabel("Category")  # X-axis label
ax[0, 0].set_ylabel("Revenue $")  # Y-axis label

cos = df.groupby('item_type', observed=False)['cost'].mean()
cos = rev.sort_values(ascending=False)
cos.plot(kind="bar", ax=ax[0, 1], color="red")
#df.groupby("item_type", observed=False)["cost"].mean().plot(kind="bar", ax=ax[0, 1], color="red")
ax[0, 1].set_title("Cost by Category")  # Title of the subplot
ax[0, 1].set_xlabel("Category")  # X-axis label
ax[0, 1].set_ylabel("Cost $")  # Y-axis label

pro = df.groupby('item_type', observed=False)['profit'].mean()
pro = rev.sort_values(ascending=False)
pro.plot(kind="bar", ax=ax[1, 0], color="green")
#df.groupby("item_type", observed=False)["profit"].mean().plot(kind="bar", ax=ax[1, 0], color="green")
ax[1, 0].set_title("Profit by Category")  # Title of the subplot
ax[1, 0].set_xlabel("Category")  # X-axis label
ax[1, 0].set_ylabel("Profit $")  # Y-axis label

mar = df.groupby('item_type', observed=False)['margin'].mean()
mar = rev.sort_values(ascending=False)
mar.plot(kind="bar", ax=ax[1, 1], color="purple")
#df.groupby("item_type", observed=False)["margin"].mean().plot(kind="bar", ax=ax[1, 1],  color="purple")
ax[1, 1].set_title("Margin by Category")  # Title of the subplot
ax[1, 1].set_xlabel("Category")  # X-axis label
ax[1, 1].set_ylabel("Margin %")  # Y-axis label


plt.tight_layout()


plt.show()
```

```python
df_country = df.groupby('country_code', as_index=False).agg(
revenue=('revenue','sum'),
units_sold=('units_sold','sum'),
profit=('profit','sum'),
orders=('order_id','nunique')
)
df_country.sort_values('revenue', ascending=False).head()

fig = px.choropleth( df_country,
locations='country_code',            # або 'country_name'
color='revenue',
hover_name='country_code',          # або 'country_name'
hover_data=['revenue','units_sold','profit'],
color_continuous_scale='Blues',
projection='natural earth',         # або 'equirectangular'
title='Revenue by country'
)
fig.update_layout(coloraxis_colorbar=dict(title='Revenue (USD)'))
fig.show()
```

Найбільші продажі відбуваються в період з суботи по неділю, в проміжок вівторок - четвер маємо невелику просадку через сплив робочого тижня.

```python
top_countries = df.groupby('name')['revenue'].sum().nlargest(5).index

df_trends = df[df['name'].isin(top_countries)].groupby(['order_month', 'name'])['revenue'].sum().reset_index()
df_trends['order_month'] = df_trends['order_month'].dt.to_timestamp()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_trends, x='order_month', y='revenue', hue='name')
plt.title('Тренди виручки по місяцях (Топ-5 країн)')
plt.xticks(rotation=45)
plt.show()
```
